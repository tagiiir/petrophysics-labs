```python
import numpy as np 
import pandas as pd
from matplotlib import pyplot as plt
from numpy import linalg as la
```

```python
K = pd.read_csv("/Users/tagirgalimov/Desktop/ЕРЭ-петрофиз/k.mdb.csv", sep=' ', header=None, skiprows=lambda x: x % 2 == 0)
K = K.drop(K.columns[[0, 1, 2, 3, 4, 5, -1]], axis=1)
K
```

```python
U = pd.read_csv("/Users/tagirgalimov/Desktop/ЕРЭ-петрофиз/u.mdb.csv", sep=' ', header=None, skiprows=lambda x: x % 2 == 0)
U = U.drop(U.columns[[0, 1, 2, 3, 4, 5, -1]], axis=1)
U
```

```python
Th = pd.read_csv("/Users/tagirgalimov/Desktop/ЕРЭ-петрофиз/th.mdb.csv", sep=' ', header=None, skiprows=lambda x: x % 2 == 0)
Th = Th.drop(Th.columns[[0, 1, 2, 3, 4, 5, -1]], axis=1)
Th
```

```python
K_sum = K.sum()
K_sum
```

```python
U_sum = U.sum()
Th_sum = Th.sum()
```

```python
Spectr = pd.DataFrame ({
    'K_sum': K_sum,  
    'U_sum': U_sum,  
    'Th_sum': Th_sum 
}) 
Spectr.reset_index(drop=True, inplace=True)
Spectr.index += 1 
Spectr
```

```python
plt.figure(figsize=(12, 6))
plt.plot(Spectr.index, Spectr['K_sum'], label='K_sum')
plt.plot(Spectr.index, Spectr['U_sum'], label='U_sum')
plt.plot(Spectr.index, Spectr['Th_sum'], label='Th_sum')

plt.yscale('log')
plt.xlabel('Номер спектра')
plt.ylabel('Сумма по спектру')
plt.title('Спектры по номерам')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
quest = pd.read_csv("/Users/tagirgalimov/Desktop/ЕРЭ-петрофиз/quest.mdb.csv", sep=' ', header=None, skiprows=lambda x: x % 2 == 0)
quest = quest.drop(quest.columns[[0, 1, 2, 3, 4, 5, -1]], axis=1)
quest = quest.T
quest
```

```python
C = np.linalg.pinv(Spectr) @ quest
C.index = ('K', 'U', 'Th')

```

```python
C_df = pd.DataFrame(C, index=['K', 'U', 'Th'])

plt.figure(figsize=(12, 6))

plt.plot(C_df.columns, C_df.loc['K'], label='K (Калий)', color='green')
plt.plot(C_df.columns, C_df.loc['U'], label='U (Уран)', color='blue')
plt.plot(C_df.columns, C_df.loc['Th'], label='Th (Торий)', color='red')

plt.xlabel('Номер измерения / столбца в Quest')
plt.ylabel('Значения коэффициентов')
plt.title('Коэффициенты для Калия, Урана и Тория')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
length_sample_cm = 12  
interval_cm = 5
num_samples = 15
sample_indices = np.linspace(0, C_df.shape[1] - 1, num_samples, dtype=int)

sequence = []
    
for i in sample_indices:
    k_val = C_df.loc['K', C_df.columns[i]]
    u_val = C_df.loc['U', C_df.columns[i]]
    th_val = C_df.loc['Th', C_df.columns[i]]
        

    max_element = np.argmax([k_val, u_val, th_val])
    if max_element == 0:
        sequence.append('K')
    elif max_element == 1:
        sequence.append('U')
    elif max_element == 2:
        sequence.append('Th')

print(sequence)
```
