```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
file_path2 = '/Users/tagirgalimov/Desktop/Петрофизика-2сем/Таёжное 102.csv'
file_path3 = '/Users/tagirgalimov/Desktop/Петрофизика-2сем/Таёжное 103.csv'
```

```python
t2 = pd.read_csv(file_path2, sep = ';')
t3 = pd.read_csv(file_path3, sep = ';')
t2
```

```python
t2 = t2.fillna('')
t3 = t3.fillna('')
t2
```

```python
t2 = t2[::-1]
index_2 = range(1,14,1)
t2.index = index_2
t3 = t3[::-1]
index_3 = range(1,17,1)
t3.index = index_3
t3
```

```python
rocks_t2 = t2['Rock'] + t2['Color'] + ' ' + t2['ТЕКСТУРА'] + ' ' + t2['Включения органогенные'] + ' ' + t2['Включения минеральные'] + ' ' + t2['НЕФТЕНАСЫЩЕНИЕ'] + ' ' + t2['ПОДОШВА СЛОЯ']
rocks_t3 = t3['Rock'] + t3['Color'] + ' ' + t3['ТЕКСТУРА'] + ' ' + t3['Включения органогенные'] + ' ' + t3['Включения минеральные'] + ' ' + t3['НЕФТЕНАСЫЩЕНИЕ'] + ' ' + t3['ПОДОШВА СЛОЯ']
rocks_t2
```

```python
rocks_2_list = [rock for rock in rocks_t2]
rocks_3_list = [rock for rock in rocks_t3]
levensht_matrix = np.zeros((len(rocks_3_list)+1, len(rocks_2_list)+1))
levensht_matrix
```

```python
levensht_matrix[0, :] = np.arange(len(rocks_2_list)+1)
levensht_matrix[:, 0] = np.arange(len(rocks_3_list)+1)
levensht_matrix
```

```python
for i in range(1, (len(rocks_3_list)+1)):
    for j in range(1, (len(rocks_2_list)+1)):
        up_value = levensht_matrix[i-1, j] + 1
        left_value = levensht_matrix[i, j-1] + 1
        
        if rocks_2_list[j-1] == rocks_3_list[i-1]:
            diag_value = levensht_matrix[i-1, j-1]
        else:
            diag_value = levensht_matrix[i-1, j-1] + 1
        
        value = min([left_value, up_value, diag_value])
        levensht_matrix[i,j] = value
        
levensht_matrix
```

```python
plt.figure(figsize=(10, 7))
plt.imshow(levensht_matrix, cmap='viridis') 
plt.colorbar(label='Value')
for i in range(levensht_matrix.shape[0]):
    for j in range(levensht_matrix.shape[1]):
        plt.text(j, i, f"{int(levensht_matrix[i, j])}", ha='center', va='center', color='white')
plt.title('Heatmap (imshow)')
plt.xlabel('Столбцы')
plt.ylabel('Строки')
plt.show()
```

```python
def trace_path(matrix, list1, list2):
    path = []
    i, j = matrix.shape[0]-1, matrix.shape[1]-1
    correlations = []

    while i > 0 or j > 0:
        path.append((i, j))
        if i == 0:
            correlations.append(f"Пласт {j-1} из списка 2 выклинивается")
            j -= 1
        elif j == 0:
            correlations.append(f"Пласт {i-1} из списка 3 выклинивается")
            i -= 1
        else:
            min_val = min(matrix[i-1, j-1], matrix[i-1, j], matrix[i, j-1])
            if matrix[i-1, j-1] == min_val:
                if list1[j-1] == list2[i-1]:
                    correlations.append(f"Пласты {j-1} и {i-1} совпадают")
                else:
                    correlations.append(f"Пласты {j-1} и {i-1} частично совпадают")
                i -= 1
                j -= 1
            elif matrix[i-1, j] == min_val:
                correlations.append(f"Пласт {i-1} из списка 3 выклинивается")
                i -= 1
            else:
                correlations.append(f"Пласт {j-1} из списка 2 выклинивается")
                j -= 1

    path.append((0, 0))
    return path[::-1], correlations[::-1]
```

```python
path, correlations = trace_path(levensht_matrix, rocks_2_list, rocks_3_list)

plt.figure(figsize=(10, 7))
plt.imshow(levensht_matrix, cmap='viridis')
plt.colorbar(label='Расстояние')
plt.title('Матрица Левенштейна')

for i in range(levensht_matrix.shape[0]):
    for j in range(levensht_matrix.shape[1]):
        plt.text(j, i, f"{int(levensht_matrix[i, j])}", ha='center', va='center', color='white')

for (i1, j1), (i2, j2) in zip(path[:-1], path[1:]):
    plt.plot([j1, j2], [i1, i2], 'r-', linewidth=2)

plt.xlabel("rocks_2_list")
plt.ylabel("rocks_3_list")
plt.show()

print("Результаты корреляции:")
for line in correlations:
    print(line)
```

```python
def prepare_well_data(depths, rocks, heights):
    new_depth = []
    rock_plot = []
    codes = {'П': 3, 'Г': 2, 'У': 1}

    for i in range(len(depths)):
        d = depths[i]
        h = heights[i]
        r = rocks[i]
        code = codes.get(r[0], 4)

        if i == 0:
            new_depth += [d, d]
        elif i == len(depths) - 1:
            new_depth += [d] * 4
            new_depth += [d + h] * 2
        else:
            new_depth += [d] * 4

        rock_plot += [0, code, code, 0]

    return [-x for x in new_depth], rock_plot

new_depth_2, rock_2_plot = prepare_well_data(
    list(t2['Depth']),
    list(t2['Rock']),
    list(t2['H'])
)

new_depth_3, rock_3_plot = prepare_well_data(
    list(t3['Depth']),
    list(t3['Rock']),
    list(t3['H'])
)

rock_3_plot = [-x for x in rock_3_plot]

import matplotlib.pyplot as plt

fig, ax = plt.subplots(ncols=2, figsize=(15, 10))

ax[0].plot(rock_2_plot, new_depth_2, c='r', lw=2, marker='o')
ax[1].plot(rock_3_plot, new_depth_3, c='b', lw=2, marker='o')

ax[0].set_title('Таёжное 102')
ax[1].set_title('Таёжное 103')
ax[0].set_xlabel('Породы')
ax[1].set_xlabel('Породы')
ax[0].set_ylabel('Глубина')
ax[1].set_ylabel('Глубина')

plt.tight_layout()
plt.show()
```
